```python
from threading import *
from time import *
class Hello(Thread):
    def run(self):
        for i in range(100):
             
            print("Hello There!\n")
            sleep(1)
            
        
class Hey(Thread):
    def run(self):
        for i in range(100):
            print("Hey There\n")
            sleep(1)
```

```python
t1=Hello()
t2=Hey()
```

```python
t1.start()
sleep(0.2)
t2.start()
```
